```python
#import libraries 
import cv2 as cv ##open CV 
from PIL import Image ##pillow library 
import numpy as np


##helper fxns
import importlib as il
import demoutils as utils
il.reload(utils)
```

```python
#read image
demoImg = cv.imread("./data/img2.png")
utils.display(demoImg)
```

```python
#convert to grayscale 
grayImg= cv.cvtColor(demoImg,cv.COLOR_BGR2GRAY)
utils.display(grayImg)
```

```python
#resize image 
resizedImg = cv.resize(grayImg, (500,500))
utils.display(resizedImg)

##advanced extension: image interpolation
```

```python
#normalize pixel intensities 
normalizedImg = cv.normalize(grayImg, None, alpha=0, beta=255, norm_type=cv.NORM_MINMAX)
utils.display(normalizedImg)
##advanced extension: Histogram Equalization
```

```python
#binarization 
_,finalImg = cv.threshold(normalizedImg,210,255,cv.THRESH_BINARY_INV)
utils.display(finalImg)

##advanced extension: adaptive thresholding 
```

```python
#denoise 
k = np.ones((3, 3), np.uint8)

erode = cv.erode(finalImg, k, iterations=2) 
denoisedImg = cv.dilate(erode, k, iterations=2)  

utils.display(denoisedImg)
utils.runDenoise(normalizedImg)

##advanced extension: bilateral, gaussian and mean image filtering 
```
